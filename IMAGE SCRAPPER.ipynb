```python
import requests
import logging
import os
from bs4 import BeautifulSoup # scrap data from html/xml pages
from urllib.request import urlopen  #request from a particular server -> urllib
```

```python
#create a directory where all images will be saved, if it doesnt exsits in my system
save_dir="images/"
if not os.path.exists(save_dir):     #if this pasth doesnt exists, 
    os.makedirs(save_dir)            #create this path
```

```python
query="Virat Kohli"
#url= https://www.google.fr/search?sca_esv=570462611&q=elon+musk&tbm=isch&source=lnms&sa=X&ved=2ahUKEwikn5m8hduBAxXC0QIHHXbVB7cQ0pQJegQICRAB&biw=1440&bih=724&dpr=2
```

```python
response=requests.get(f"https://www.google.fr/search?sca_esv=570462611&q={query}&tbm=isch&source=lnms&sa=X&ved=2ahUKEwikn5m8hduBAxXC0QIHHXbVB7cQ0pQJegQICRAB&biw=1440&bih=724&dpr=2")
```

```python
#from the response above, we will fetch the content
soup=BeautifulSoup(response.content, "html.parser") # working with html content, like we get on inspect search for anything on browser(cleint) and request on server
```

```python
soup # we created a vrariasoup.findble and got the html content.
```

```python
# now to get all the images
image_tags=soup.find_all("img")
```
